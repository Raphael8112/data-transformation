```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Titanic Dataset.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
#countplot for survived
sns.countplot(x='Survived',data=data)
```

```python
#count of gender that survived
sns.countplot(x='Survived',hue='Gender',data=data)
#
```

```python
#plot for embarked
sns.countplot(x='Embarked',data=data)
```

```python
#divide the categorical columns into noninal
noninal=['Name','Ticket','Cabin']
ordinal_cat=['Embarked','Gender']
```

```python
gender_categories=['Gender','Male']
embarked_categories=['t','Q','S']
#convert the data into the categories
data['Gender']=pd.Categorical(data['Gender'],categories=gender_categories,ordered=False)
data['Embarked']=pd.Categorical(data['Embarked'],categories=embarked_categories,ordered=False)
#
```

```python
#counterplot of gender
sns.countplot(x='Gender',data=data)
```
